```python
import pandas as pd

# This path matches the folder structure in your screenshots
file_path = r"C:\Users\LENOVO\Desktop\Canaan internship\Week 1\sales_data\train.csv"

try:
    # 1. Load the data
    df = pd.read_csv(file_path, encoding="latin1")
    print("Success! Data loaded.")
    
    # 2. Remove Duplicates
    rows_before = len(df)
    df.drop_duplicates(inplace=True)
    print(f"Duplicates removed: {rows_before - len(df)}")

    # 3. Handle Missing Values
    df = df.fillna(0)
    print("Missing values handled.")

    # 4. Format Columns
    df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
    print("Columns formatted.")

    # 5. Save the Cleaned File
    save_path = r"C:\Users\LENOVO\Desktop\Canaan internship\Week 1\cleaned_sales_data.csv"
    df.to_csv(save_path, index=False)
    print(f"Task 1 Complete! Cleaned file saved at: {save_path}")

    # Show the first few rows
    display(df.head())

except Exception as e:
    print(f"Error: {e}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# --- Task 2: Exploratory Data Analysis (EDA) ---

# 1. Summary Statistics
print("Summary Statistics:")
print(df.describe())

# 2. Check Data Types and Nulls
print("\nDataset Info:")
print(df.info())

# 3. Visualization: Top 10 States by Count
plt.figure(figsize=(10,5))
df['state'].value_counts().head(10).plot(kind='bar', color='salmon')
plt.title('Top 10 States in Dataset')
plt.xlabel('State')
plt.ylabel('Number of Orders')
plt.show()

# 4. Visualization: Ship Mode Distribution
plt.figure(figsize=(6,6))
df['ship_mode'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Shipping Modes')
plt.ylabel('')
plt.show()
```
